## v. Outlier Handling

```python
def calculate_outliers(df):
    """
    Fungsi untuk menghitung total baris dan persentase baris dalam DataFrame yang memiliki outlier.
    
    Parameters:
    df (pd.DataFrame): DataFrame yang akan dianalisis.

    Returns:
    int: Total baris dengan outlier.
    float: Persentase baris yang memiliki outlier.
    """
    
    # Menghitung total baris dalam dataset
    total_rows = len(df)

    # Membuat kolom untuk menandai baris yang memiliki outlier
    has_outlier = pd.Series([False] * total_rows)

    # Menghitung skewness dan menentukan outlier untuk setiap kolom numerik
    for column in df.select_dtypes(include=['float64', 'int64']).columns:
        skewness = df[column].skew()

        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Tentukan batas outlier berdasarkan skewness
        if skewness < -1 or skewness > 1:
            lower_bound = Q1 - 3 * IQR
            upper_bound = Q3 + 3 * IQR
        else:
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

        # Tandai baris yang memiliki outlier di kolom ini
        has_outlier |= (df[column] < lower_bound) | (df[column] > upper_bound)

    # Hitung jumlah baris yang memiliki setidaknya satu outlier
    total_outlier_rows = has_outlier.sum()

    # Menghitung persentase baris yang memiliki outlier
    outlier_percentage = (total_outlier_rows / total_rows) * 100

    return total_outlier_rows, outlier_percentage
```

```python
# Pilih kolom yang ingin di cek outliernya
df_eda_num = df_eda[['Age', 'Date.diff']]

# masukan kedalam fungsi pengecekan outlier
# Contoh penggunaan
total_outliers1, percentage_outliers1 = calculate_outliers(df_eda_num)
print(f"Total Rows with Outliers in data: {total_outliers1}")
print(f"Percentage of Rows with Outliers in data: {percentage_outliers1:.2f}%")
```

> **Insight**

- **Identifikasi Outlier**: Dalam dataset pelatihan (`X_train`), terdapat **2.408 baris yang teridentifikasi sebagai outlier**, yang setara dengan **2.25% dari total baris**. 
- **Prosentase Outlier yang Rendah**: Meskipun ada outlier, persentasenya relatif kecil (hanya 2.25%). Ini menunjukkan bahwa sebagian besar data berada dalam rentang yang diharapkan, sehingga outlier ini mungkin tidak mencerminkan pola umum dalam dataset.
- **Strategi Penanganan Outlier**: Mengingat jumlah outlier yang relatif kecil, capping dapat menjadi pendekatan yang baik untuk menanganinya. Dengan capping, kita dapat mengubah nilai outlier ke batas tertentu, yang dapat membantu dalam mempertahankan informasi dari data tanpa mengabaikannya secara keseluruhan. 
- **Menjaga Kualitas Data**: Capping akan membantu menjaga integritas dan kualitas data, serta mencegah outlier mempengaruhi hasil analisis dan model secara signifikan, terutama untuk model yang sensitif terhadap nilai ekstrim.

Dengan demikian, pendekatan capping untuk mengatasi outlier adalah langkah yang tepat untuk memastikan model yang lebih robust dan akurat.


```python
def cap_outliers(df, columns):
    """
    Fungsi untuk melakukan capping pada outlier di kolom yang ditargetkan berdasarkan skewness dan IQR.
    
    Parameters:
    df (pd.DataFrame): DataFrame yang akan diproses.
    columns (list): Daftar nama kolom yang ingin dilakukan capping.

    Returns:
    pd.DataFrame: DataFrame dengan outlier yang telah dicapping.
    """
    
    # Looping untuk setiap kolom yang ditargetkan
    for column in columns:
        if column in df.columns:
            # Mengubah kolom menjadi float untuk mencegah masalah dtype
            df[column] = df[column].astype(float)
            
            # Hitung skewness
            skewness_value = df[column].skew()

            # Menghitung batas bawah dan atas untuk outlier
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            # Tentukan batas outlier berdasarkan skewness
            if skewness_value < -1 or skewness_value > 1:
                lower_bound = Q1 - 3 * IQR
                upper_bound = Q3 + 3 * IQR
            else:
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR

            # Lakukan capping dengan .loc[] untuk menghindari SettingWithCopyWarning
            df.loc[df[column] < lower_bound, column] = lower_bound
            df.loc[df[column] > upper_bound, column] = upper_bound
        else:
            print(f"Kolom {column} tidak ditemukan di DataFrame.")

    return df
```

> **Insight**

  - **Age**: Nilai-nilai ekstrem pada kolom usia telah dicapping untuk memastikan bahwa analisis tidak terpengaruh oleh data yang tidak representatif.
  - **Date.diff**: Proses capping juga diterapkan pada kolom perbedaan tanggal untuk menjaga agar nilai-nilai tetap relevan dan tidak mengganggu analisis lebih lanjut.

- **Tujuan Capping**: 
  - Capping outliers bertujuan untuk mengurangi dampak negatif dari nilai ekstrem pada model pembelajaran mesin, sehingga meningkatkan stabilitas dan akurasi hasil model.

- **Persiapan Data**: 
  - Dengan melakukan capping, dataset `df_eda` kini lebih bersih dan siap untuk analisis selanjutnya, seperti visualisasi atau pelatihan model, dengan mengandalkan data yang lebih konsisten dan representatif.

### i. Bagaimana distribusi usia pasien (`Age`) yang tidak hadir dibandingkan dengan yang hadir pada janji medis?

```python
# Membuat fungsi untuk mengelompokkan usia dengan nama yang lebih deskriptif
def categorize_age(age):
    if age <= 12:
        return 'Anak-anak (0-12)'
    elif 13 <= age <= 19:
        return 'Remaja (13-19)'
    elif 20 <= age <= 35:
        return 'Dewasa Muda (20-35)'
    elif 36 <= age <= 55:
        return 'Dewasa (36-55)'
    else:
        return 'Lanjut Usia (56+)'

# Menerapkan fungsi untuk membuat kolom kategori usia baru
df_eda['Age_Group'] = df_eda['Age'].apply(categorize_age)

# Menghitung jumlah pasien berdasarkan kelompok usia dan kehadiran
age_group_counts = df_eda.groupby(['Age_Group', 'Showed_up']).size().unstack(fill_value=0)

# Menambahkan kolom jumlah tidak hadir untuk pengurutan
age_group_counts['Jumlah_Tidak_Hadir'] = age_group_counts[False]

# Mengurutkan berdasarkan jumlah tidak hadir, menurunkan
sorted_age_groups = age_group_counts.sort_values(by='Jumlah_Tidak_Hadir', ascending=False).index

# Plot distribusi kategori usia pasien yang hadir dan tidak hadir
plt.figure(figsize=(10, 6))
sns.countplot(data=df_eda, x='Age_Group', hue='Showed_up', order=sorted_age_groups, palette='Set2')

# Menambahkan judul dan label
plt.title('Distribusi Kategori Usia Pasien Berdasarkan Kehadiran pada Janji Medis', fontsize=14)
plt.xlabel('Kelompok Usia', fontsize=12)
plt.ylabel('Jumlah Pasien', fontsize=12)

# Menambahkan legend untuk penjelasan rentang umur
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Tidak Hadir', 'Hadir'], title='Kehadiran')

# Menampilkan plot
plt.show()
```

> **Insight**

Grafik "Distribusi Kategori Usia Pasien Berdasarkan Kehadiran pada Janji Medis" menunjukkan jumlah pasien dari berbagai kelompok usia yang datang (hadir) dan tidak datang (tidak hadir) pada janji medis mereka.

- **Kelompok Usia Dewasa (36-55)**: Memiliki jumlah pasien yang paling banyak, baik yang hadir maupun tidak hadir. Ini mengindikasikan bahwa kelompok usia ini merupakan kelompok pasien terbesar yang berkunjung ke fasilitas medis tersebut.
- **Tingkat Ketidakhadiran**: Jumlah pasien yang tidak hadir cenderung lebih tinggi dibandingkan dengan yang hadir pada semua kelompok usia. Hal ini menunjukkan bahwa masih banyak pasien yang tidak memenuhi janji medis mereka, terlepas dari kelompok usianya.
- **Kelompok Usia Anak dan Remaja (0-19)**: Proporsi pasien yang tidak hadir cenderung lebih tinggi pada kelompok usia anak-anak (0-12) dan remaja (13-19). Ini mungkin disebabkan oleh beberapa faktor, seperti jadwal kegiatan yang padat, kurangnya kesadaran akan pentingnya janji medis, atau faktor lain yang terkait dengan usia.
- **Kelompok Usia Lanjut (56+)**: Memiliki jumlah pasien yang tidak hadir relatif lebih sedikit dibandingkan dengan kelompok usia dewasa muda (20-35). Ini bisa jadi karena kelompok usia lanjut lebih sadar akan pentingnya kesehatan dan cenderung lebih patuh pada jadwal pengobatan.
  
Secara keseluruhan, grafik ini menunjukkan bahwa tingkat kehadiran pada janji medis masih menjadi tantangan, terutama pada kelompok usia anak-anak dan remaja.

## ii. Cardinality Check

```python
def determine_cardinality(df):
    """
    Menghitung cardinality dari setiap kolom dalam DataFrame.

    Args:
        df (pd.DataFrame): DataFrame yang akan dianalisis.

    Returns:
        pd.DataFrame: DataFrame berisi nama kolom, jumlah nilai unik, dan nilai unik untuk setiap kolom.
    """
    # Membuat DataFrame untuk menyimpan hasil cardinality
    cardinality_list = []

    # Mencetak cardinality dalam format tabel
    for column in df.columns:
        unique_values = df[column].unique()  # Mengambil nilai unik
        unique_count = len(unique_values)     # Menghitung jumlah nilai unik
        cardinality_list.append({
            'Nama Kolom': column,
            'Jumlah Unique': unique_count,
            'Unique Values': unique_values
        })

    # Mengubah list hasil cardinality menjadi DataFrame
    cardinality_df = pd.DataFrame(cardinality_list)

    return cardinality_df
```

```python
# 2. Identifikasi kolom numerik untuk capping outlier
columns_numerical = X_train.select_dtypes(include=['float64', 'int64']).columns.tolist()

# 3. Fungsi untuk melakukan capping pada outlier
def cap_outliers(df, columns):
    for column in columns:
        if column in df.columns:
            df[column] = df[column].astype(float)
            skewness_value = df[column].skew()
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            
            if skewness_value < -1 or skewness_value > 1:
                lower_bound = Q1 - 3 * IQR
                upper_bound = Q3 + 3 * IQR
            else:
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR
            
            # Lakukan capping
            df.loc[df[column] < lower_bound, column] = lower_bound
            df.loc[df[column] > upper_bound, column] = upper_bound
        else:
            print(f"Kolom {column} tidak ditemukan di DataFrame.")
    return df

# 4. Capping outliers di data pelatihan
X_train_num = cap_outliers(X_train.copy(), columns_numerical)

# 5. Menentukan batas untuk capping berdasarkan data pelatihan
lower_bounds = {}
upper_bounds = {}
for column in columns_numerical:
    Q1 = X_train_num[column].quantile(0.25)
    Q3 = X_train_num[column].quantile(0.75)
    IQR = Q3 - Q1
    if X_train_num[column].skew() < -1 or X_train_num[column].skew() > 1:
        lower_bounds[column] = Q1 - 3 * IQR
        upper_bounds[column] = Q3 + 3 * IQR
    else:
        lower_bounds[column] = Q1 - 1.5 * IQR
        upper_bounds[column] = Q3 + 1.5 * IQR

# 6. Terapkan capping ke data pengujian dengan batas dari data pelatihan
X_test_num = X_test[columns_numerical].copy()
for column in columns_numerical:
    X_test_num[column] = X_test_num[column].clip(lower=lower_bounds[column], upper=upper_bounds[column])

# 7. Update X_train dan X_test dengan data numerik yang telah dicapping
X_train[columns_numerical] = X_train_num[columns_numerical]
X_test[columns_numerical] = X_test_num[columns_numerical]

# 8. Pisahkan kolom kategorikal
X_train_cat = X_train.select_dtypes(include=['object', 'bool'])
X_test_cat = X_test.select_dtypes(include=['object', 'bool'])
# 9. pisahkan kolom numerik lagi
X_train_num = X_train.select_dtypes(include=['float64', 'int64'])
X_test_num = X_test.select_dtypes(include=['float64', 'int64'])


# Sekarang X_train dan X_test telah siap digunakan untuk modeling
print("Data Pelatihan dan Pengujian telah siap.")
print("X_train_num:", X_train_num.head())
print("X_train_cat:", X_train_cat.head())
print("X_test_num:", X_test_num.head())
print("X_test_cat:", X_test_cat.head())
```

```python
import pandas as pd
from scipy import stats

def check_cor_ken(data, target):
    """
    Memeriksa korelasi antara fitur kategorikal dan target.

    Args:
        data (pd.DataFrame): DataFrame berisi fitur kategorikal.
        target (pd.Series atau pd.DataFrame): Target yang akan dibandingkan dengan fitur.

    Returns:
        pd.DataFrame: DataFrame yang berisi fitur, metode, nilai korelasi, p-value, dan hasil korelasi.
    """
    # Mengubah target menjadi Series jika merupakan DataFrame
    target_series = target.squeeze() if isinstance(target, pd.DataFrame) else target
    results = []

    for i in data.columns:
        # Mengubah nilai kategorikal menjadi kode numerik
        numeric_values = data[i].astype('category').cat.codes
        
        # Menghitung Kendall's Tau
        cor_value, p_val = stats.kendalltau(numeric_values, target_series)

        # Menentukan kekuatan korelasi
        if cor_value > 0.5:
            correlation = "Strong Positive"
        elif 0 < cor_value <= 0.5:
            correlation = "Weak Positive"
        elif cor_value == 0:
            correlation = "No Correlation"
        elif -0.5 < cor_value < 0:
            correlation = "Weak Negative"
        elif cor_value < -0.5:
            correlation = "Strong Negative"
        else:
            correlation = "Nan"

        results.append([i, 'Kendall', cor_value, p_val, correlation])

    results_df = pd.DataFrame(results, columns=['Feature', 'Method', 'Cor-Value', 'P-Value', 'Result'])
    return results_df
```

```python
def check_cor_chi2(data, target):
    """
    Memeriksa korelasi antara fitur kategorikal dan target menggunakan uji Chi-Squared.

    Args:
        data (pd.DataFrame): DataFrame berisi fitur kategorikal.
        target (pd.Series atau pd.DataFrame): Target yang akan dibandingkan dengan fitur.

    Returns:
        pd.DataFrame: DataFrame yang berisi fitur, metode, nilai chi-squared, p-value, dan hasil korelasi.
    """
    # Mengubah target menjadi Series jika merupakan DataFrame
    target_series = target.squeeze() if isinstance(target, pd.DataFrame) else target
    results = []

    for feature in data.columns:
        # Menghitung tabel kontingensi antara fitur dan target
        contingency_table = pd.crosstab(data[feature], target_series)
        
        # Melakukan uji Chi-Squared
        chi2_stat, p_val, _, _ = chi2_contingency(contingency_table)

        # Menentukan apakah hasilnya signifikan
        if p_val < 0.05:
            significance = "Significant"
        else:
            significance = "Not Significant"

        results.append([feature, 'Chi-Squared', chi2_stat, p_val, significance])

    results_df = pd.DataFrame(results, columns=['Feature', 'Method', 'Cor/F-Value', 'P-Value', 'Result'])
    return results_df
```

Metrik Evaluasi Model

Sebelum kita membahas hasilnya, berikut adalah beberapa metrik evaluasi yang digunakan:

- **Precision**: Proporsi prediksi positif yang benar dari seluruh prediksi positif. Precision penting ketika kita ingin meminimalkan False Positives (kesalahan dalam memprediksi kehadiran ketika pasien tidak hadir).
  - Formula: Precision = TP / (TP + FP)
  
- **Recall**: Proporsi prediksi positif yang benar dari seluruh data positif yang sebenarnya. Recall lebih relevan saat kita ingin meminimalkan False Negatives (kesalahan dalam memprediksi ketidakhadiran ketika pasien hadir).
  - Formula: Recall = TP / (TP + FN)
  
- **F1-Score**: Rata-rata harmonis dari Precision dan Recall. F1-Score memberikan keseimbangan antara Precision dan Recall.
  - Formula: F1-Score = 2 * (Precision * Recall) / (Precision + Recall)
  
- **Accuracy**: Proporsi prediksi yang benar terhadap total data. Accuracy dapat menyesatkan ketika dataset tidak seimbang.
  - Formula: Accuracy = (TP + TN) / (TP + FP + TN + FN)
  
- **ROC-AUC (Area Under Curve)**: Mengukur kemampuan model dalam memisahkan kelas positif dan negatif. Nilai ROC-AUC berkisar antara 0 hingga 1, di mana nilai lebih dekat ke 1 berarti performa lebih baik.
  - Evaluasi ROC-AUC:
    - Excellent: 0.90 - 1.00
    - Good: 0.80 - 0.90
    - Fair: 0.70 - 0.80
    - Poor: 0.60 - 0.70
    - Fail: 0.50 - 0.60

```python
def evaluate_roc_auc(roc_auc_value):
    """Evaluasi skor ROC-AUC ke dalam kategori evaluasi"""
    if roc_auc_value >= 0.9:
        return "Excellent"
    elif roc_auc_value >= 0.8:
        return "Good"
    elif roc_auc_value >= 0.7:
        return "Fair"
    elif roc_auc_value >= 0.6:
        return "Poor"
    else:
        return "Fail"

def evaluate_model(model, X_train, X_test, y_train, y_test):
    """
    Evaluasi model dengan ROC-AUC, plot ROC Curve, dan classification report.

    Args:
        model: Model yang telah dilatih.
        X_train: Data pelatihan.
        X_test: Data pengujian.
        y_train: Target pelatihan.
        y_test: Target pengujian.
    """
    # Menghitung prediksi probabilitas untuk kelas positif
    y_train_prob = model.predict_proba(X_train)[:, 1]
    y_test_prob = model.predict_proba(X_test)[:, 1]

    # Menghitung ROC-AUC
    roc_auc_train = roc_auc_score(y_train, y_train_prob)
    roc_auc_test = roc_auc_score(y_test, y_test_prob)

    # Menampilkan hasil evaluasi ROC-AUC
    print('ROC-AUC - Train : ', roc_auc_train)
    print('ROC-AUC - Test  : ', roc_auc_test)

    print("ROC-AUC Train Evaluation: ", evaluate_roc_auc(roc_auc_train))
    print("ROC-AUC Test Evaluation: ", evaluate_roc_auc(roc_auc_test))

    # Menghitung dan menampilkan ROC Curve
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)

    # Plot ROC Curve
    plt.figure(figsize=(10, 6))
    plt.plot(fpr_train, tpr_train, color='blue', lw=2, label='ROC Curve Train (AUC = {:.2f})'.format(roc_auc_train))
    plt.plot(fpr_test, tpr_test, color='orange', lw=2, label='ROC Curve Test (AUC = {:.2f})'.format(roc_auc_test))
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)  # Garis diagonal
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')
    plt.grid()
    plt.show()

    # Menampilkan Classification Report
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    print("Classification Report - Train Data")
    print(classification_report(y_train, y_train_pred))

    print("Classification Report - Test Data")
    print(classification_report(y_test, y_test_pred))

    # Evaluasi beberapa metrics
    def evaluate_model_performance(y_true, y_pred, y_prob):
        acc = accuracy_score(y_true, y_pred)
        precision = precision_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred)
        roc_auc = roc_auc_score(y_true, y_prob)

        print(f"Accuracy       : {acc:.2f}")
        print(f"Precision      : {precision:.2f}")
        print(f"Recall         : {recall:.2f}")
        print(f"F1-Score       : {f1:.2f}")
        print(f"ROC-AUC        : {roc_auc:.2f}")
        print(f"ROC-AUC Evaluation : {evaluate_roc_auc(roc_auc)}")

    # Evaluasi performa pada train data
    print("Evaluation Metrics - Train Data")
    evaluate_model_performance(y_train, y_train_pred, y_train_prob)

    # Evaluasi performa pada test data
    print("Evaluation Metrics - Test Data")
    evaluate_model_performance(y_test, y_test_pred, y_test_prob)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, classification_report, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

def evaluate_svm_model(model, X_train, X_test, y_train, y_test):
    """
    Evaluasi model SVM dengan ROC Curve, dan classification report.

    Args:
        model: Model yang telah dilatih.
        X_train: Data pelatihan.
        X_test: Data pengujian.
        y_train: Target pelatihan.
        y_test: Target pengujian.
    """
    # Menghitung prediksi kelas
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Metrik evaluasi
    print("Classification Report - Train Data")
    print(classification_report(y_train, y_train_pred))

    print("Classification Report - Test Data")
    print(classification_report(y_test, y_test_pred))

    # Evaluasi beberapa metrics
    def evaluate_model_performance(y_true, y_pred):
        acc = accuracy_score(y_true, y_pred)
        precision = precision_score(y_true, y_pred)
        recall = recall_score(y_true, y_pred)
        f1 = f1_score(y_true, y_pred)

        print(f"Accuracy       : {acc:.2f}")
        print(f"Precision      : {precision:.2f}")
        print(f"Recall         : {recall:.2f}")
        print(f"F1-Score       : {f1:.2f}")

    # Evaluasi performa pada train data
    print("Evaluation Metrics - Train Data")
    evaluate_model_performance(y_train, y_train_pred)

    # Evaluasi performa pada test data
    print("Evaluation Metrics - Test Data")
    evaluate_model_performance(y_test, y_test_pred)

    # ROC Curve (tanpa probabilitas)
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_pred)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_pred)

    # Plot ROC Curve
    plt.figure(figsize=(10, 6))
    plt.plot(fpr_train, tpr_train, color='blue', lw=2, label='ROC Curve Train')
    plt.plot(fpr_test, tpr_test, color='orange', lw=2, label='ROC Curve Test')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)  # Garis diagonal
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (Tanpa Probabilitas)')
    plt.legend(loc='lower right')
    plt.grid()
    plt.show()
```

```python
%time
# Evaluasi SVM
evaluate_svm_model(pipeline_svm, X_train, X_test, y_train, y_test)
```

# xii. Conceptual Problem

> **1.Latar Belakang dan Cara Kerja Bagging**

**Latar Belakang Bagging:**
Bagging, atau Bootstrap Aggregating, adalah teknik ensemble yang digunakan untuk meningkatkan stabilitas dan akurasi model machine learning. Konsep ini diperkenalkan untuk mengurangi varians dari model dengan menggabungkan beberapa model yang lemah (weak learners) menjadi satu model yang kuat (strong learner). Bagging sangat berguna dalam situasi di mana model yang digunakan cenderung overfitting pada data pelatihan.

**Cara Kerja Bagging:**
1. **Sampling:** Dalam bagging, beberapa subset dari data pelatihan diambil dengan teknik bootstrap, yaitu dengan pengambilan sampel secara acak dengan pengembalian. Setiap subset ini akan digunakan untuk melatih model yang terpisah.
  
2. **Model Training:** Model yang sama (misalnya, decision tree) dilatih pada setiap subset data yang dihasilkan. Karena setiap model dilatih pada subset data yang berbeda, mereka akan memiliki variasi dalam prediksi mereka.

3. **Aggregasi:** Setelah model dilatih, prediksi dari setiap model digabungkan. Untuk model klasifikasi, biasanya digunakan voting mayoritas untuk menentukan kelas akhir, sementara untuk regresi, rata-rata dari prediksi semua model diambil.

> **2.Perbedaan Cara Kerja Random Forest dengan Algoritma Boosting**

**Random Forest:**
- Random Forest adalah teknik ensemble yang menggunakan bagging untuk menggabungkan banyak decision tree. 
- Setiap tree dibangun berdasarkan subset data yang diambil secara acak, dan pada setiap split, hanya subset acak dari fitur yang dipertimbangkan. 
- Hasil dari semua pohon diambil dengan metode voting mayoritas untuk klasifikasi atau rata-rata untuk regresi.

**Boosting:**
- Boosting adalah teknik ensemble yang bekerja dengan cara melatih model secara berurutan. 
- Model pertama dilatih pada seluruh data, dan model berikutnya dilatih dengan memberikan bobot lebih pada data yang salah diklasifikasikan oleh model sebelumnya. 
- Dengan cara ini, boosting berfokus untuk memperbaiki kesalahan dari model-model sebelumnya, sehingga meningkatkan akurasi secara bertahap. Contoh algoritma boosting adalah AdaBoost dan Gradient Boosting.

> **3.Apa yang Dimaksud dengan Cross Validation**

**Cross Validation:**
Cross Validation adalah teknik yang digunakan untuk mengevaluasi kemampuan generalisasi dari model machine learning. Dengan metode ini, dataset dibagi menjadi beberapa subset (folds). Model dilatih pada beberapa fold dan diuji pada fold yang tersisa. Proses ini diulang beberapa kali, dengan setiap fold bertindak sebagai data uji sekali. Hasil evaluasi dari setiap iterasi kemudian dirata-ratakan untuk memberikan estimasi kinerja model yang lebih akurat. Cross Validation membantu dalam menghindari overfitting dan memberikan gambaran yang lebih baik tentang bagaimana model akan bekerja pada data yang tidak terlihat. Metode yang paling umum digunakan adalah K-Fold Cross Validation.
